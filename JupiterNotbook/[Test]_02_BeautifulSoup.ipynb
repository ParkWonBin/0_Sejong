````python
# requests 부분
import requests               
url = "http://info.hanatour.com/?hanacode=main_TNB_getabout"
html = requests.get(url).text
````

 ## BeautifulSoup 사용하기
 
 **요약 :**
전체 웹 코드에서 원하는(특정) **부분을 추출**하기 위해 사용하는 함수

**설치 :**
**cmd** 창에서 **pip install bs4** 입력

````python
from bs4 import BeautifulSoup

# BeautifulSoup로 소스 받아서 lxml로 번역( parsing)
soup = BeautifulSoup(html, "lxml")
# 첫번째 인자 : html 소스코드
# 두번째 인자 : parser 선택
````

### 1. Parsing
parser 는 번역한다는 뜻이다.    
lxml을 잘 아는 전문가를 대려와서 html을 해독하게 한다.

````python
# 출력 : 태그 전체
print(soup)
````

## 2. Select Data by Tag
태그로 특정 부분 데이터만 추출 :   
**soup.select(" 태그명 ")**

<img src="img/bs1.png">

````python
soup.select("title")
````

## 3. Print Data
태그 내용만 보기 : .text   
태그 속성만 보기 : .get("src")

**3.1 반복문**

````python
data = soup.select("title")
for i in  data :
    print(i.text)
````

**3.2 리스트 요소 선택**

````python
soup.select("title")[0].text

# soup.select("title").text
# select 반환값이 리스트이기 떄문에 바로 .text할 수는 없다.
````

## 4. Select Data by CSS
CSS 선택자로 특정 부분 데이터만 추출 :   
**soup.select(" CSS 선택자 ")**

**html 구조**
```html
<div class="menu">
    <a class="depth1" href="/dest/">인기여행도시</a>
    <div class="depth2 dest">
        <div class="column">
            <strong class="title">동남아</strong>
            <ul>
                <li><a href="/dest/city/17">세부</a></li>
                <li><a href="/dest/city/16">푸껫</a></li>
                <li><a href="/dest/city/44">다낭</a></li>
                <li><a href="/dest/city/52">후아힌</a></li>
                <li><a href="/dest/city/53">나트랑</a></li>
                <li><a href="/dest/city/54">씨엠립</a></li>
                <li><a href="/dest/city/18">보라카이</a></li>
                <li><a href="/dest/city/41">마닐라</a></li>
            </ul>
        </div>
    </div>
</div>

```

<img src="img/cssSelectors.png">

````python
# 자손(하위) 선택자 :
data = soup.select(".menu ul li")

# 자식(후손) 선택자 : 
# data = soup.select(".menu > ul > li")
````

[자손 vs 자식 선택자](https://jw910911.tistory.com/32)   

**자손(하위) 선택자 :** 자식,손자, 후손을 모두 선택된다.   

**자식(후손) 선택자 :** 특정 요소의 직계 자식만 선택한다.

````python
for i in  data :
    print(i.text)
````
