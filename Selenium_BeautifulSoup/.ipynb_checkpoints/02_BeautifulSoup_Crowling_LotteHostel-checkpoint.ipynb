```python
# 롯데호텔 데이터 추출하기

# 셀레늄 설치 : cmd 창에서 pip install selenium
# 뷰티플 스프 설치 : cmd 창에서 pip install bs4
# 크롬 드라이버 설치 :  https://sites.google.com/a/chromium.org/chromedriver/downloads
# SessionNotCreatedException : [크롬 - 더보기 - 도움말 - chrome 정보] 들어가서 브라우저 버전확인
```

```python
# === [Part 1 : 라이블러리 참조 및 브라우저 열기] ======
from selenium import webdriver # 봇으로 페이지 들어가기
from bs4 import BeautifulSoup  # 페이지 특정 부분 추출하기

# 크롬 브라우저 열기
driver = webdriver.Chrome('C:/chromedriver.exe')

# 롯데호텔 룸 프로모션 화면으로 이동
url = "https://www.lottehotel.com/global/ko/hotel-offers/room-promotions.html"
driver.get(url)
```

```python
# === [Part 2 : 데이터 추출 및 출력] ======

# BeautifulSoup로 소스 받아오기
html = driver.page_source
soup = BeautifulSoup(html, "lxml")

# 데이터 선택해서 리스트로 저장
promotion__title = soup.select('.promotion__title')
promotion__date = soup.select('.promotion__date')
promotion_price__number = soup.select('.promotion-price__number')

# 저장한 데이터 출력
for i in range(len(promotion__title)):
    print('제목 :', promotion__title[i].text)
    print('날짜 :', promotion__date[i].text.strip().lstrip('예약 ,투숙 '))
    print('가격 :', promotion_price__number[i].text)   
    print('\n')
```

```python
# === [비교 화면] ======

# Beautifulsoup & request로 페이지 가져오면 JS 코드가 그대로 넘어온다. 
# Selenium으로 페이지를 로드한 후에 페이지 소스를 로드해야 내용물을 얻을 수있다..

import requests
from bs4 import BeautifulSoup

url = "https://www.lottehotel.com/global/ko/hotel-offers/room-promotions.html"
html = requests.get(url).text
soup = BeautifulSoup(html, "lxml")

# 데이터 선택해서 리스트로 저장
promotion__title = soup.select('.promotion__title')
promotion__date = soup.select('.promotion__date')
promotion_price__number = soup.select('.promotion-price__number')

# 저장한 데이터 출력
for i in range(len(promotion__title)):
    print('제목 :', promotion__title[i].text)
    print('날짜 :', promotion__date[i].text.strip())
    print('가격 :', promotion_price__number[i].text)   
    print('\n')
```
