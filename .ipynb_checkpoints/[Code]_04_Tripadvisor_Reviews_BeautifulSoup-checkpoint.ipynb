#### 전체코드 

````python
from selenium import webdriver
from bs4 import BeautifulSoup 

# 크롬 브라우저 열기
driver = webdriver.Chrome('C:/chromedriver.exe')

# 트립어드바이저 리뷰 화면으로 이동
url = "https://www.tripadvisor.co.kr/Hotel_Review-g294197-d306130-Reviews-Lotte_Hotel_Seoul-Seoul.html"
driver.get(url)

ReviewID,ReviewTitle,ReviewText,ReviewDate = [],[],[],[]

page = 1               
# 1 ~ page 번째 페이지까지 읽어옴
for i in range(page) :
    
    # BeautifulSoup로 소스 받아오기
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")
    
    # 리뷰글 : 작성자, 글제목, 내용, 숙박일
    ReviewID   .extend(soup.select('.social-member-event-MemberEventOnObjectBlock__event_type--3njyv span a'))
    ReviewTitle.extend(soup.select('.location-review-review-list-parts-ReviewTitle__reviewTitleText--2tFRT'))
    ReviewText .extend(soup.select('.location-review-review-list-parts-ExpandableReview__reviewText--gOmRC'))
    ReviewDate .extend(soup.select('.location-review-review-list-parts-EventDate__event_date--1epHa'))
    
    # 다음 페이지로 이동 버튼 클릭
    selector = ".ui_button.nav.next.primary "
    driver.find_element_by_css_selector(selector).click()
    
# 저장된 정보 출력
for i in range(len(ReviewID)):
    print(" #",i+1,':', ReviewDate[i].text.lstrip('숙박 날짜 :')) 
    print('이름 :', ReviewID[i].text) 
    print('제목 :', ReviewTitle[i].text)
    print('내용 :', ReviewText[i].text,'\n')   
````

# 문제 및 해결

    pageNum = 2

    url_2 = '-Lotte_Hotel_Seoul-Seoul.html#REVIEWS'
    url = url_1+str((pageNum-1)*5)+url_2
    driver.get(url)
```

### 2. 중복데이터 저장

문제 : 트립어드바이저에서 "다음"버튼을 눌렀을 떄 가끔씩 제대로 이동되지 않음.  <br>
해결1 : 게시글 작성자의 이름을 비교하여, 중복데이터가 들어오면 페이지 로드 재시도.  <br>
해결2 : "다음"버튼 누르는 시간이 1.5초보다 짧으면 버튼이 제대로 작동되지 않는 것 확인,   <br>

````python
from selenium import webdriver
from bs4 import BeautifulSoup 
import time

# 크롬 브라우저 열기
driver = webdriver.Chrome('C:/chromedriver.exe')

# 트립어드바이저 리뷰 화면으로 이동
url = "https://www.tripadvisor.co.kr/Hotel_Review-g294197-d306130-Reviews-Lotte_Hotel_Seoul-Seoul.html"
driver.get(url)

# 변수 초기화
page = 5
temp = ""
nextBtn = ".ui_button.nav.next.primary "
ReviewID,ReviewTitle,ReviewText,ReviewDate = [],[],[],[]

# 크롤링 반복
current_page = 0
while current_page < page :
    
    # 페이지 로드 후 게시글 작성자 확인
    soup = BeautifulSoup(driver.page_source, "lxml")
    checkID = soup.select('.social-member-event-MemberEventOnObjectBlock__event_type--3njyv span a')[0].text
    
    # 게시글 달라지면 저장
    if temp != checkID :                      
        
        # 리뷰글 : 작성자, 글제목, 내용, 숙박일
        ReviewID   .extend(soup.select('.social-member-event-MemberEventOnObjectBlock__event_type--3njyv span a'))
        ReviewTitle.extend(soup.select('.location-review-review-list-parts-ReviewTitle__reviewTitleText--2tFRT'))
        ReviewText .extend(soup.select('.location-review-review-list-parts-ExpandableReview__reviewText--gOmRC'))
        ReviewDate .extend(soup.select('.location-review-review-list-parts-EventDate__event_date--1epHa'))
    
        temp = checkID      # 현제 페이지 정보 갱신
        current_page += 1   # 현제 페이지 번호 갱신
    
    else :
        print("클릭 실패")
        time.sleep(1.5) # 게시글 이동 실패하면 1.5초 후 재시도
        
    # 다음 페이지 버튼 클릭
    driver.find_element_by_css_selector(nextBtn).click()
    
# 저장된 정보 출력
for i in range(len(ReviewID)):
    print(" #",i+1,':', ReviewDate[i].text.lstrip('숙박 날짜 :')) 
    print('이름 :', ReviewID[i].text) 
    print('제목 :', ReviewTitle[i].text)
    print('내용 :', ReviewText[i].text,'\n')   
````

# 엑셀로 저장하기

````python
import os
import pandas as pd

SaveDate,SaveID,SaveTitle,SaveText = [],[],[],[]

for i in range(len(ReviewID)):
    SaveDate  .append( ReviewDate[i] .text.lstrip('숙박 날짜 :'))
    SaveID    .append( ReviewID[i]   .text)
    SaveTitle .append( ReviewTitle[i].text)
    SaveText  .append( ReviewText[i] .text)
      
df = pd.DataFrame({ '날짜': SaveDate, '이름': SaveID, '제목': SaveTitle,'내용': SaveText})

print(df)

User = os.path.expanduser("~")+"/Desktop"
forderName = "TripAdviser"
forderURL = f"{User}/{forderName}/"

#해당 경로에 폴더가 없으면 폴더 생성 (import os)
if not os.path.exists(forderURL): os.makedirs(forderURL)   
    
# pandas data frame 을 csv로 저장하는 명령어
df.to_csv(forderURL+"Tripadviser.csv")
````

# 함수형으로 다시 정리

### User Defined Function

함수를 사용하는 이유 :     
이 글에선 반복되는 코드를 정리하는 용도로 사용한다.

**함수 만드는 방법 :**
```python
def 함수이름(입력값1,입력값2 ...) :
    실행할 내용```
    
**함수 호출 :**
```python
함수이름(입력값1,입력값2 ...)```

링크 바로가기 : [파이썬 함수 설명글](https://wikidocs.net/24#_1)

````python
from selenium import webdriver
from bs4 import BeautifulSoup 
import time

# 로드된 페이지 읽고 데이터 추가

def LoadHtml() : 
    global soup,html
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")

def AddData() :
    global ReviewID,ReviewTitle,ReviewText,ReviewDate
    ReviewID    += soup.select('.social-member-event-MemberEventOnObjectBlock__event_type--3njyv span a')
    ReviewTitle += soup.select('.location-review-review-list-parts-ReviewTitle__reviewTitleText--2tFRT')
    ReviewText  += soup.select('.location-review-review-list-parts-ExpandableReview__reviewText--gOmRC')
    ReviewDate  += soup.select('.location-review-review-list-parts-EventDate__event_date--1epHa')

# 데이터 리셋    
def ResetData() :
    # 1페이지로 이동
    url = "https://www.tripadvisor.co.kr/Hotel_Review-g294197-d306130-Reviews-Lotte_Hotel_Seoul-Seoul.html"
    driver.get(url)
      
# 다음 페이지 버튼 클릭
def NextPage() :                               
    
    # "다음 페이지로 이동" 버튼을 클릭한다.
    selector = ".ui_button.nav.next.primary "
    driver.find_element_by_css_selector(selector).click()
    
#저장된 자료 출력
def PrintData() :
    for i in range(len(ReviewID)):
        print(" #",i+1,':', ReviewDate[i].text.lstrip('숙박 날짜 :')) 
        print('이름 :', ReviewID[i].text) 
        print('제목 :', ReviewTitle[i].text)
        print('내용 :', ReviewText[i].text,'\n')    
    
def CheckLoad() : 
    global tempID,checkID
    checkID = soup.select('.social-member-event-MemberEventOnObjectBlock__event_type--3njyv span a')[0].text
    print(f"탬프 : {tempID} \n체크 : {checkID}\n")
    if tempID == checkID :
        print("=== [중복] ===\n")
        NextPage()
        time.sleep(3)
        LoadHtml()
        CheckLoad()
    else :
         tempID = checkID
    


# 트립어드바이저 페이지로 이동
driver = webdriver.Chrome('C:/chromedriver.exe')
url = "https://www.tripadvisor.co.kr/Hotel_Review-g294197-d306130-Reviews-Lotte_Hotel_Seoul-Seoul.html"
driver.get(url)

ReviewID,ReviewTitle,ReviewText,ReviewDate,tempID = [],[],[],[],""

# 함수 호출
ResetData()              # 초기화
page = 5                 # 리뷰글 페이지
for i in range(page) :  # 1 ~ 5 페이지까지 읽어옴
    LoadHtml()           # 페이지 읽기 시도
    CheckLoad()          # 같은 페이지 읽었으면 다음버튼, 아니면 그냥 진행 
    AddData()            # 데이터 저장
    NextPage()           # 다음 페이지로 이동
PrintData()              # 출력
````
